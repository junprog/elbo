```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
```

```python
map_2d = np.zeros((50,50))
map_2d[9,19] = 1
map_2d[26,22] = 1
map_2d[40,10] = 1
```

```python
plt.figure()
plt.imshow(map_2d)
plt.colorbar()
plt.show()
```

```python
dense_map = gaussian_filter(map_2d, 4)
```

```python
plt.figure()
plt.imshow(dense_map)
plt.colorbar()
plt.show()
```

```python
pri_prob = dense_map / np.sum(dense_map)
```

```python
plt.figure()
plt.imshow(pri_prob)
plt.colorbar()
plt.show()

np.sum(pri_prob)
```

```python
pri_prob += 1e-10 
E_nt = (-1) * pri_prob * np.log(pri_prob)
```

```python
plt.figure()
plt.imshow(E_nt)
plt.colorbar()
plt.show()
```

```python
np.sum(E_nt)
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
soft_val = softmax(dense_map)
```

```python
plt.figure()
plt.imshow(soft_val)
plt.colorbar()
plt.show()
```

```python
np.sum(soft_val)
```

```python
soft_E_nt = (-1)*soft_val * np.log(soft_val)
```

```python
plt.figure()
plt.imshow(soft_E_nt)
plt.colorbar()
plt.show()
```

```python
np.sum(soft_E_nt)
```

```python
E = np.matmul(soft_val, dense_map)
```

```python
np.sum(E)
```
